```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
csv_file = "US Energy Sources By Type 1973 to 2019.csv"
energy_sources_raw_df = pd.read_csv(csv_file)
energy_sources_raw_df.head()
```

```python
energy_sources_raw_df.dtypes
```

```python
energy_sources_raw_df
```

```python
energy_sources_raw_df.set_index("Value", inplace=True)
energy_sources_raw_df.head()
```

```python
energy_sources_raw_df = energy_sources_raw_df.drop('Not Available', axis=0)
```

```python
energy_sources_raw_df.reset_index(inplace=True)
energy_sources_raw_df.head()
```

```python
energy_sources_raw_df['Value'] = energy_sources_raw_df.Value.astype(float)
```

```python
energy_sources_raw_df.dtypes
```

```python
energy_sources_raw_df['year'] = energy_sources_raw_df['YYYYMM'].floordiv(100)
energy_sources_raw_df.head()
```

```python
energy_sources_group_df = energy_sources_raw_df.groupby(['Description','year']).sum()
energy_sources_group_df.head()
```

```python
energy_sources_group_df.reset_index(inplace=True)
energy_sources_group_df.head()
```

```python
energy_sources_group_df = energy_sources_group_df[['Description','year','Value']]
energy_sources_group_df.head()
```

```python
is_total_ff = energy_sources_group_df['Description'] == 'Total Fossil Fuels Production'
energy_ff = energy_sources_group_df[is_total_ff]
energy_ff.set_index("year", inplace=True)
energy_ff.rename(columns = {'Value':'fossil_fuels_qbtu'}, inplace = True) 
energy_ff
```

```python
is_total_nuclear = energy_sources_group_df['Description'] == 'Nuclear Electric Power Production'
energy_nuclear = energy_sources_group_df[is_total_nuclear]
energy_nuclear.set_index("year", inplace=True)
energy_nuclear.rename(columns = {'Value':'nuclear_qbtu'}, inplace = True) 
energy_nuclear
```

```python
is_total_renewable = energy_sources_group_df['Description'] == 'Total Renewable Energy Production'
energy_renewable = energy_sources_group_df[is_total_renewable]
energy_renewable.set_index("year", inplace=True)
energy_renewable.rename(columns = {'Value':'renewable_qbtu'}, inplace = True)
energy_renewable
```

```python
is_total_total = energy_sources_group_df['Description'] == 'Total Primary Energy Production'
energy_total = energy_sources_group_df[is_total_total]
energy_total.set_index("year", inplace=True)
energy_total.rename(columns = {'Value':'total_energy_production_qbtu'}, inplace = True)
energy_total
```

```python
energy_sources_df = energy_ff[['fossil_fuels_qbtu']]
energy_sources_df['nuclear_qbtu'] = energy_nuclear['nuclear_qbtu']
energy_sources_df['renewable_qbtu'] = energy_renewable['renewable_qbtu']
energy_sources_df['total_energy_production_qbtu'] = energy_total['total_energy_production_qbtu']
energy_sources_df
```

```python
# Greenhouse Data
greenhouse_file = "US Greenhouse Gas Emissions By Sector 1990 to 2017.csv"
greenhouse_raw_df = pd.read_csv(greenhouse_file, thousands=',')
greenhouse_raw_df.head()
```

```python
# Set Index to 'year'
greenhouse_raw_df.set_index("Economic Sector", inplace=True)
greenhouse_raw_df.head()
```

```python
# Transpose rows to columns 
greenhouse_transposed = greenhouse_raw_df.T
greenhouse_transposed.head()
```

```python
emissions_df = greenhouse_transposed[['Electricity generation']]
emissions_df.head()
```

```python
emissions_df.reset_index(inplace=True)
emissions_df.head()
```

```python
emissions_df.rename(columns = {'Electricity generation':'electricity_emissions',
                              'index':'year'
                              }, inplace = True)
emissions_df.head()
```

```python
emissions_df.set_index('year', inplace=True, drop=True)
emissions_df.head()
```

```python
# Population Data
population_file = "US Total Population 1960 to 2018.csv"
population_df = pd.read_csv(population_file, thousands=',')
population_df.head()
```

```python
# Set Index to 'year'
population_df.set_index('Country Name', inplace=True)
population_df.head()
```

```python
# Remove Country Code, Indicator Name and Indicator Code
population_df.drop(['Country Code','Indicator Name','Indicator Code'],axis = 1,inplace = True)
population_df.head()
```

```python
# Transpose rows to columns 
population_transposed_df = population_df.T
population_transposed_df.head()
```

```python
# Parse out United States 
US_population_all_df = population_transposed_df [['United States']]
US_population_all_df.head()
```

```python
US_population_all_df.reset_index(drop=False, inplace=True)
US_population_all_df.head()
```

```python
# Rename United States as us_population
US_population_all_df.rename(columns = {'United States':'us_population',
                                      'index':'year'
                                      }, inplace = True)
US_population_all_df.head()
```

```python
US_population_all_df.set_index('year', inplace=True)
US_population_all_df.head()
```

```python
US_population_all_df.dropna()
US_population_all_df.head()
```

```python
emissions_population_df = pd.merge(emissions_df, US_population_all_df, left_index=True, right_index=True)
emissions_population_df
```

```python
energy_sources_df
```

```python
emissions_population_df.reset_index(inplace=True)
```

```python
emissions_population_df['year'] = emissions_population_df.year.astype(int)
emissions_population_df.set_index('year', inplace=True)
```

```python
power_emissions_population_df = pd.merge(energy_sources_df, emissions_population_df, left_index=True, right_index=True)
power_emissions_population_df
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/project_2_etl"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
power_emissions_population_df.to_sql(name='renewable_energy_db', con=engine, if_exists='append', index=True)
```
