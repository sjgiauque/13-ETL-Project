```python
import pandas as pd
from sqlalchemy import create_engine
```

### Extract CSVs into DataFrames

```python
# Greenhouse Data
greenhouse_file = "../Resources/USCO2Emissions19902017.csv"
greenhouse_df = pd.read_csv(greenhouse_file)
greenhouse_df.head()
```

```python
# Rename the column headers
greenhouse_df.rename(columns = {'Economic Sector/Year':'year'}, inplace = True)
greenhouse_df.head()
```

```python
# Set Index to 'year'
greenhouse_df.set_index("year", inplace=True)
greenhouse_df.head()
```

```python
# Population Data
population_file = "../Resources/USPopulation19602018.csv"
population_df = pd.read_csv(population_file)
population_df.head()
```

```python
# Rename the column headers
population_df.rename(columns = {'Country Name':'year'}, inplace = True)
population_df.head()
```

```python
# Set Index to 'year'
population_df.set_index("year", inplace=True)
population_df.head()
```

### Transform Greenhouse DataFrame

```python
# Transform rows to columns 
greenhouse_transposed = greenhouse_df.T
greenhouse_transposed.head()
```

### Transform Population DataFrame

```python
# Remove Country Code, Indicator Name and Indicator Code
population_df.drop(['Country Code','Indicator Name','Indicator Code'],axis = 1,inplace = True)
population_df.head()
```

```python
# Transform rows to columns 
population_transposed_df = population_df.T
population_transposed_df.head()
```

```python
# Parse out United States 
US_population_df = population_transposed_df [['United States']]
US_population_df.head()
```

```python
# Rename United States as us_population
US_population_df.rename(columns = {'United States':'us_population'}, inplace = True)
US_population_df.head()
```

### Create database connection

```python
connection_string = "postgres:postgres@localhost:5432/Greenhouse_Emissions"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

### Load DataFrames into database

```python
greenhouse_transposed.to_sql(name='greenhouse', con=engine, if_exists='append', index=True)
```

```python
US_population_df.to_sql(name='population', con=engine, if_exists='append', index=True)
```
